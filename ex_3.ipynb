```python
import pandas as pd;
from apyori import apriori;
```

```python
data = pd.read_csv('example.csv',header=None);

records = []
for i in range(0,data.shape[1]):
    records.append(list(data[i].dropna()))
results = list(apriori(records,min_support=0.22, min_length=2));
```

```python
for item in results:
    items = [x for x in item[0]]
    pairs = list(item[2]);
    del pairs[0];
    for p in pairs:
        print("Rule: ", ", ".join(list(p[0])), " -> ", ", ".join(list(p[1])));
        print("Support: ",  str(item[1]))
        print("Confidence: ", list(p)[2])
        print("Lift: ", list(p)[3])
        print("=====================================")
```
